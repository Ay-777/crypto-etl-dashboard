```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/processed/crypto_clean.csv")
plt.bar(df["id"], df["current_price"])
plt.title("Current Crypto Prices (USD)")
plt.show()
```

```python
plt.bar(df["id"], df["current_price"])
plt.yscale('log')
plt.title("Crypto Prices (Log Scale)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
bars = plt.bar(df["id"], df["current_price"], color='skyblue')
plt.title("Current Crypto Prices (USD)")
plt.xlabel("Cryptocurrency")
plt.ylabel("Price (USD)")
plt.grid(axis='y')

# Add price labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"${height:,.0f}", ha='center', va='bottom')

plt.show()
```
